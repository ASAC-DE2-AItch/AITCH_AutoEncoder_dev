# ae_v1 — v1 집계 MLP-AE + 참조 베이스라인 + 6층 하네스  (P2·P3·P4)

Autoencoder 개발계획 **v3 §6(피처)·§7(하네스)·§8(모델링)** 실행 노트북. 입력 = `merged_data_v3.parquet`.
**주력 아키텍처(판단 v1)**: 신호는 시간축이 아니라 **레벨축**(정착 T=2) → v1 **집계 MLP-AE + 과도 모양 지표**. 시퀀스 AE는 백로그.

### 이 노트북이 하는 것 (위→아래 순차)
1. **P0** 환경·시드·상수·로딩(+**dedup 가드**·split 드랍, 불가침 11) — `ae_eda_v3` 유틸 복제
2. **P0.5** 세트 경계 확정 + **해시 대조**(ae_eda_v3 재현성 — §8 열린확인 (b) 채택: 전체 19게이트 재실행 대신 해시 대조)
3. **P2** 피처 명세(feature_spec) + wafer×피처 추출(과도 모양 지표 포함) + winsorize/Robust(fit=TRAIN만)
4. **P3** 분할(TRAIN/VAL/seg2·C6_1 격리) + 합성 주입(부록 B) + **6층 평가 하네스**
5. **P4** 베이스라인(PCA·Mahalanobis) → MLP-AE 학습 → 스코어(ae_raw/top_channels/drift) → **시드 3회 안정성** → **G1 판정**

### 불가침 (코드로 강제)
- **1** C65 입력금지·C64 그룹분할 · **2** seg2·C6_1·INJECT-EVAL은 G판정 셀에서만 로딩 · **3** 스케일러 fit=TRAIN만
- **5** 과도(C42=1)/정착(C42=0) 분리 · **7** 물리신호만(부록 A) · **8** 출력 비움·사용자 실행 · **10** 모순 시 중단·보고
- **11** 로드 시 dedup 가드 `(C64,C7,C46)` + split 드랍 상시 · **12** 최종출력 단일 `ae_raw` 계약 불변

> ⚠️ seg2·C6_1·INJECT-EVAL은 §7-3 오염방지: **G 판정 셀(맨 끝 근처)에서만 로딩**. HPO 목적함수 = VAL 재구성 loss만.


```python
# ── P0-1 · 환경·의존성 체크 (계획서 §4) ───────────────────────────────
# 목적: 런타임 고정·기록. 이 노트북(ae_v1~)이 계획서 §4 P0 GPU DoD 확인 지점.
import sys, platform, hashlib, warnings, json
from pathlib import Path
import numpy as np
import pandas as pd
warnings.filterwarnings('ignore')
try:
    import IPython.display as _d
    def show(x):
        try: _d.display(x)
        except Exception: print(x.to_string() if hasattr(x,'to_string') else x)
except Exception:
    def show(x): print(x.to_string() if hasattr(x,'to_string') else x)

_env = {'python': platform.python_version(), 'numpy': np.__version__, 'pandas': pd.__version__}
for mod in ['pyarrow','sklearn','optuna','mlflow']:
    try:
        m = __import__(mod); _env[mod] = getattr(m,'__version__','?')
    except Exception as e:
        _env[mod] = f'MISSING ({type(e).__name__})'
try:
    import torch
    _env['torch'] = torch.__version__
    _env['cuda_available'] = bool(torch.cuda.is_available())
    _env['cuda_device'] = torch.cuda.get_device_name(0) if torch.cuda.is_available() else 'CPU-fallback'
except Exception as e:
    _env['torch'] = f'MISSING ({type(e).__name__})'; _env['cuda_available'] = False

print('── 런타임 스냅샷 (REPORT 기록용) ──')
for k, v in _env.items(): print(f'  {k:16s}: {v}')
assert sys.version_info >= (3, 10), 'python >= 3.10 필요 (계획서 §4)'
if 'MISSING' in str(_env['torch']):
    print('\n[중단 후보] torch 미설치 — 모델링 노트북은 torch 필수. requirements.txt로 설치 후 재실행.')
elif not _env.get('cuda_available'):
    print('\n[주의] CUDA 미인식 — CPU 폴백. AE 규모는 CPU도 수 분(N-2 재학습 예산과 동일 경로). GPU는 HPO 폭 확대용.')
else:
    print('\n[OK] GPU 인식 — 계획서 §4 P0 GPU DoD 충족 (모델링 노트북 확인 지점).')
```

```python
# ── P0-2 · 재현성 시드 + 프로젝트 상수 (계획서 §4, 불가침 1·7 · 부록 A 단일소스) ──
import random
SEED = 42
def seed_all(s=SEED):
    random.seed(s); np.random.seed(s)
    try:
        import torch; torch.manual_seed(s)
        if torch.cuda.is_available(): torch.cuda.manual_seed_all(s)
        try: torch.use_deterministic_algorithms(True)
        except Exception:
            torch.backends.cudnn.deterministic = True; torch.backends.cudnn.benchmark = False
    except Exception: pass
seed_all(SEED)
print(f'seed={SEED} 고정 (random/numpy/torch)')

def _find(name):
    for p in [Path(name), Path('..')/name, Path('../..')/name]:
        if p.exists(): return p.resolve()
    return (Path('..')/name).resolve()
CSV_PATH = _find('merged_data_v3.csv')
PARQUET_PATH = CSV_PATH.with_suffix('.parquet')
ART_DIR = Path('artifacts'); ART_DIR.mkdir(exist_ok=True)   # ae_v1/artifacts 아님 — 노트북 폴더 하위
print(f'입력: {PARQUET_PATH.name}  · 아티팩트: {ART_DIR.resolve()}')

# 컬럼 그룹 (데이터 사전 §6 · 부록 A)
STR_COLS    = ['C6','C14','C20','C21','C22','C23','C24','C28','C29','C30','C38','C64','split']
CORRUPT_COL = 'C40'
DROP_EMPTY  = ['C2','C13','C26','C37','C43','C47','C53','C55']
DROP_DUP    = ['C36','C35','C38','C40']
CONST_COLS  = ['C3','C8','C14','C19','C21','C24','C28','C29','C30','C44','C45','C51']
GROUP_KEYS  = ['C24','C6','C7']
FORBIDDEN_INPUT = set(['C65']
    + ['C64','C38','C20','C21','C22','C23','C34','C35','C14','C24','C6','C7','C36']
    + ['C10','C39','C40','C41','C46','C33'] + ['C1','C4','C5','C48'] + ['C12'] + ['C42'] + ['split'])

# 부록 A 포함 채널 (물리 신호)
CH_SETTLE_CONT = ['C11','C17','C9','C52','C15','C16','C31','C63','C58','C57']
CH_TRANSIENT   = ['C18','C27','C32','C62','C61']
CH_DISCRETE    = ['C49','C54','C56','C50']
CH_MUX         = ['C59','C60']

# 채널 손실 가중 w_c (부록 A 초안 — P4 학습곡선으로 확정)
CH_LOSS_W = {'C11':1.0,'C17':1.0,'C9':1.0,'C52':0.7,'C15':1.0,'C16':1.0,'C31':1.0,'C63':0.7,
             'C58':0.5,'C57':0.3,'C18':1.0,'C27':0.7,'C32':1.0,'C62':1.0,'C61':0.7,
             'C54':0.7,'C56':0.7,'C50':0.5,'C49':0.3,'C5960_active':0.5}

def sha1_8(items):
    h = hashlib.sha1()
    for x in sorted(map(str, items)): h.update(str(x).encode()); h.update(b'|')
    return h.hexdigest()[:8]
print(f'상수: 금지입력 {len(FORBIDDEN_INPUT)}종 · 포함채널 정착{len(CH_SETTLE_CONT)}+과도{len(CH_TRANSIENT)}+이산{len(CH_DISCRETE)}+MUX')
```

```python
# ── P0-3 · 데이터 로딩 + parquet 캐시 (계획서 §4) ─────────────────────
def load_data(force_rebuild=False):
    if PARQUET_PATH.exists() and not force_rebuild:
        df = pd.read_parquet(PARQUET_PATH); print(f'[cache] parquet 로드: {df.shape[0]:,} × {df.shape[1]}'); return df
    if not CSV_PATH.exists():
        raise FileNotFoundError(f'merged_data_v3.csv 없음 — 노트북을 AE/ae_v1/에 두고 실행. 탐색={CSV_PATH}')
    print('[build] CSV → parquet 최초 변환 중...')
    dtypes = {c:'string' for c in STR_COLS}; dtypes[CORRUPT_COL]='string'
    df = pd.read_csv(CSV_PATH, dtype=dtypes, low_memory=False)
    df.to_parquet(PARQUET_PATH, index=False); print(f'[build] 저장: {PARQUET_PATH.name} {df.shape}')
    return df
df_raw = load_data()
print(f'로드 형상: {df_raw.shape[0]:,}행 × {df_raw.shape[1]}열 · dtype:',
      {str(k):int(v) for k,v in df_raw.dtypes.astype(str).value_counts().items()})
```

```python
# ── P0-4 · 표준 전처리 유틸 (계획서 §4 · ae_eda_v3 동일 셀 복제) ──────
def standard_sort(df):
    return df.sort_values(['C10','C64','C46'], kind='mergesort').reset_index(drop=True)
def reconstruct_c59_c60(df, thr=1000):
    out = df.copy(); big = out['C59'] > thr
    out['C5960_active'] = np.where(big, out['C59'], out['C60']).astype(float)
    out['C5960_ch']     = np.where(big, 0, 1).astype('int8')
    return out
def is_step4_transient(df): return (df['C7']==4) & (df['C42']==1)
def is_step4_settle(df):    return (df['C7']==4) & (df['C42']==0)
def dedup_merge_artifact(df):
    before = len(df)
    out = (df.sort_values(['C10','C64','C46'], kind='mergesort')
             .drop_duplicates(['C64','C7','C46'], keep='first').reset_index(drop=True))
    return out, before - len(out)
def find_regime_boundary(df):
    w = (df.groupby('C64', sort=False).agg(c33_first=('C33','first'), t_min=('C10','min'))
           .reset_index().sort_values('t_min', kind='mergesort').reset_index(drop=True))
    reset_idx = np.where((w['c33_first'].diff() <= -2).values)[0]
    info = {'n_reset': int(len(reset_idx))}
    if len(reset_idx):
        b = int(reset_idx[0])
        info.update(seg2_first_wafer=str(w.loc[b,'C64']),
                    seg1_wafers=w.iloc[:b]['C64'].tolist(), seg2_wafers=w.iloc[b:]['C64'].tolist())
    return info
print('유틸: standard_sort · dedup_merge_artifact · reconstruct_c59_c60 · is_step4_* · find_regime_boundary')
```

## P0.5 · 세트 경계 확정 + 해시 대조  *(§8 열린확인 → (b) 채택)*

데이터-버전 무결성(전체 19게이트)은 `ae_eda_v3`에서 완료. P4는 매 런 **① dedup 가드 + split 드랍(불가침 11) ② 세트 경계 해시 대조(재현성)** 만 수행 — 전체 게이트 재실행은 데이터 버전이 바뀔 때만. 해시 불일치 시 불가침 10(중단·보고).


```python
# ── P0.5 · dedup 가드 + split 드랍 + 세트 경계 해시 대조 ───────────────
df = standard_sort(df_raw)
df, N_DEDUP = dedup_merge_artifact(df)                 # 불가침 11: dedup 가드
df = df.drop(columns=['split'], errors='ignore')       # 불가침 11: split 드랍
reg = find_regime_boundary(df)
SEG1 = set(reg['seg1_wafers']); SEG2 = set(reg['seg2_wafers'])
df['seg'] = np.where(df['C64'].isin(SEG1), 'seg1', 'seg2')

# 학습셋 seg1·C6_0 → TRAIN/VAL 시간순 80/20 (C64 그룹, 계획서 §6-3·§7-1)
tr_all = df[(df['seg']=='seg1') & (df['C6']=='C6_0')]
order  = tr_all.groupby('C64')['C10'].min().sort_values()
cut = int(len(order)*0.8)
TRAIN_W = set(order.index[:cut]); VAL_W = set(order.index[cut:])
C6_1_W  = set(df.loc[df['C6']=='C6_1','C64'].unique())

# 해시 대조 (ae_eda_v3 REPORT_01 확정 — 재현성 지문)
EXPECT = {'seg1':'1b142963','seg2':'0a01fcf4','train':'6ac80376','val':'fa500f3d','c6_1':'160c72e8'}
GOT = {'seg1':sha1_8(SEG1),'seg2':sha1_8(SEG2),'train':sha1_8(TRAIN_W),'val':sha1_8(VAL_W),'c6_1':sha1_8(C6_1_W)}
print('세트 경계 해시 대조 (ae_eda_v3 재현성):')
ALL_OK = True
for k in EXPECT:
    ok = EXPECT[k]==GOT[k]; ALL_OK &= ok
    print(f'  {k:6}: {GOT[k]} {"✅" if ok else "★MISMATCH exp="+EXPECT[k]}')
assert ALL_OK, '세트 경계 해시 불일치 — 재현성 깨짐 (불가침 10: 중단·보고)'
print(f'\ndedup {N_DEDUP}행 · seg1 {len(SEG1)} / seg2 {len(SEG2)} / TRAIN {len(TRAIN_W)} / VAL {len(VAL_W)} / C6_1 {len(C6_1_W)}')
print('※ seg2·C6_1 wafer 집합은 확정했으나 통계는 §P3 G판정 셀에서만 로딩 (불가침 2)')
```

## P2 · 피처 명세 + wafer×피처 추출  *(계획서 §6-2, 부록 A 단일소스)*

wafer × 집계 피처 벡터. **Step4 정착(C42=0)/과도(C42=1) 분리**(불가침 5), **물리 신호만**(불가침 7). 과도 모양 지표(판단 v1 §3-④·v3 확장) 등재.
> ⚠️ 정착 **T=2**라 정착 std는 2행 기준(집계와 정보량 준동치 — 판단 v1 §3, 정직한 한계). 피처는 **TRAIN/VAL(seg1·C6_0)만** 추출 — seg2·C6_1은 §P3 G판정 셀에서 격리 추출(불가침 2).


```python
# ── P2 · 피처 추출 함수 (계획서 §6-2 · 부록 A) ────────────────────────
def extract_features(df_all):
    # wafer × 집계 피처. df_all = dedup·정렬된 행 (원본 물리단위).
    d = reconstruct_c59_c60(df_all)
    s4 = d[d['C7'] == 4]
    settle = s4[s4['C42'] == 0]; trans = s4[s4['C42'] == 1]
    gs, gt, g4, gd = settle.groupby('C64'), trans.groupby('C64'), s4.groupby('C64'), d.groupby('C64')
    parts = []
    # 정착 연속 mean/std/min/max
    a = gs[CH_SETTLE_CONT + ['C5960_active']].agg(['mean','std','min','max'])
    a.columns = [f'{c}_s_{s}' for c, s in a.columns]; parts.append(a)
    # 점화 과도 서명 max/mean/min (과도 구간만 — 불가침 5)
    b = gt[['C18','C27','C32','C62','C61']].agg(['max','mean','min'])
    b.columns = [f'{c}_t_{s}' for c, s in b.columns]; parts.append(b)
    # 정착 잔류 (C32/C62/C61 정착 mean/std)
    c = gs[['C32','C62','C61']].agg(['mean','std'])
    c.columns = [f'{c2}_sr_{s}' for c2, s in c.columns]; parts.append(c)
    # C54/C56 매칭 위치 — 과도 wafer-mean(주, 7/13 재정정) + 정착 mean/std(보조)
    e1 = gt[['C54','C56']].mean(); e1.columns = [f'{x}_t_mean' for x in e1.columns]
    e2 = gs[['C54','C56']].agg(['mean','std']); e2.columns = [f'{c2}_s_{s}' for c2, s in e2.columns]
    parts += [e1, e2]
    # 과도 모양 지표(파생 스칼라, 판단 v1 §3-④) + 완결성
    def _slope(v):
        v = np.asarray(v, float); return float((v[-1]-v[0])/(len(v)-1)) if len(v) > 1 else 0.0
    def _mode(x):
        m = x.mode(); return float(m.iloc[0]) if len(m) else 0.0
    shape = pd.DataFrame({
        'C11_wf_min':       g4['C11'].min(),                    # 점화 스파이크 최심점(F14 레짐 프록시 — v3 +0.21σ 약함)
        'C18_t_peak':       gt['C18'].max(),                    # 매칭 과도 피크
        'C62_ignite_delta': gt['C62'].min() - gs['C62'].mean(), # 점화 반전(첫샘플−정착)
        'C50_s4_sum':       g4['C50'].sum(),                    # Endpoint 이벤트
        'C63_s_slope':      gs['C63'].agg(_slope),              # 정착 드리프트
        'C49_mode':         g4['C49'].agg(_mode),
        'C49_switch':       (g4['C49'].nunique() > 1).astype(float),
        'transient_len':    gt.size(),                          # 정착 도달 속도(과도 길이)
        'settle_len':       gs.size(),                          # 완결성
        'n_steps':          gd['C7'].nunique(),
    })
    parts.append(shape)
    F = pd.concat(parts, axis=1).reindex(gd.size().index)       # 전 wafer 인덱스
    F['has_settle']    = F['settle_len'].notna().astype(float)  # 구조결측 플래그(완결성 피처)
    F['has_transient'] = F['transient_len'].notna().astype(float)
    return F.fillna(0.0)                                        # 정착/과도 없는 wafer → 0

FEATS_TR  = extract_features(df[df['C64'].isin(TRAIN_W)])       # TRAIN만 (불가침 2·3)
FEATS_VAL = extract_features(df[df['C64'].isin(VAL_W)])
FEAT_COLS = list(FEATS_TR.columns)
print(f'피처 행렬: TRAIN {FEATS_TR.shape} · VAL {FEATS_VAL.shape} · D={len(FEAT_COLS)}')
print('  예:', FEAT_COLS[:5], '...', FEAT_COLS[-3:])
feature_spec = {'columns': FEAT_COLS, 'n_features': len(FEAT_COLS),
                'windows': {'settle':'C7==4 & C42==0', 'transient':'C7==4 & C42==1'},
                'note': 'T=2 → 정착 std 2행 기준(판단 v1 §3). seg2/C6_1 피처는 G셀에서 격리 추출.'}
(ART_DIR/'feature_spec.json').write_text(json.dumps(feature_spec, ensure_ascii=False, indent=1))
print('feature_spec.json 저장')
```

```python
# ── P2 · 클리핑(winsorize) + Robust 스케일 (fit=TRAIN만, 불가침 3) ────
from sklearn.preprocessing import RobustScaler
CLIP_LO = FEATS_TR.quantile(0.001); CLIP_HI = FEATS_TR.quantile(0.999)   # P0.1/P99.9 (TRAIN)
def _clip(F):      return F.clip(CLIP_LO, CLIP_HI, axis=1)
scaler = RobustScaler().fit(_clip(FEATS_TR))                              # median/IQR (0→1 자동보정)
def transform(F):  return scaler.transform(_clip(F.reindex(columns=FEAT_COLS, fill_value=0.0)))
X_TR  = transform(FEATS_TR); X_VAL = transform(FEATS_VAL)
print(f'스케일 fit=TRAIN({len(FEATS_TR)}) · X_TR {X_TR.shape} · X_VAL {X_VAL.shape}')

# 채널 손실 가중 벡터 (피처→채널 매핑, 부록 A) — AE MSE·ae_raw z집계에 사용
def _feat_w(col):
    for ch, w in CH_LOSS_W.items():
        if col == ch or col.startswith(ch + '_'): return w
    return 0.5                                                            # 파생·완결성
W_VEC = np.array([_feat_w(c) for c in FEAT_COLS], float)
print(f'채널 가중 W_VEC {W_VEC.shape} (mean {W_VEC.mean():.2f})')
```

## P3 · 합성 주입(부록 B) + 6층 평가 하네스  *(계획서 §7)*

**주입(부록 B)**: VAL wafer의 물리단위에 SPIKE/LEVEL/DRIFT/STUCK/NOISE/MULTI/DROP 주입 후 표준 전처리 통과(실전 경로). INJECT-TUNE/EVAL **시드 분리 2벌**(상호 오염 차단).
**하네스**: score 벡터(높을수록 이상)를 받아 L1~L6 산출 — 모델 무관 공통 자. seg2·C6_1·INJECT-EVAL은 **G 판정 셀에서만** 로딩(불가침 2·§7-3).


```python
# ── P3 · 합성 주입 (부록 B — 대표 6채널 × 7유형, 시드 분리) ───────────
INJ_CHANS = ['C17','C11','C15','C31','C63','C54']
_SIG = {ch: float(df.loc[is_step4_settle(df) & df['C64'].isin(TRAIN_W), ch].std() or 1.0) for ch in INJ_CHANS}
def inject_features(wafers, kind, k, seed):
    # wafers(VAL) 원본 행에 주입 → 피처 추출. seed로 TUNE/EVAL 분리.
    rng = np.random.default_rng(seed)
    sub = df[df['C64'].isin(wafers)].copy()
    sm = is_step4_settle(sub)
    for ch in INJ_CHANS:
        s = _SIG[ch]
        if kind == 'SPIKE':
            idx = sub[sm].groupby('C64').head(1).index; sub.loc[idx, ch] = sub.loc[idx, ch] + k*s
        elif kind == 'LEVEL':
            sub.loc[sm, ch] = sub.loc[sm, ch] + k*s
        elif kind == 'DRIFT':
            for w, g in sub[sm].groupby('C64'):
                sub.loc[g.index, ch] = sub.loc[g.index, ch].values + np.linspace(0, k*s, len(g))
        elif kind == 'STUCK':
            for w, g in sub[sm].groupby('C64'):
                sub.loc[g.index, ch] = g[ch].iloc[0]
        elif kind == 'NOISE':
            sub.loc[sm, ch] = sub.loc[sm, ch].values + rng.normal(0, s*(k-1), int(sm.sum()))
        elif kind == 'MULTI' and ch in ('C17','C11'):
            sub.loc[sm, ch] = sub.loc[sm, ch] + k*s
    if kind == 'DROP':
        di = []
        for w, g in sub[sm].groupby('C64'):
            if len(g) > 1: di += list(rng.choice(g.index, 1, replace=False))
        sub = sub.drop(index=di)
    return transform(extract_features(sub))
INJ_TYPES = {'SPIKE':[2,4,8],'LEVEL':[0.5,1,2,4],'DRIFT':[1,2],'STUCK':[0],'NOISE':[2,4],'MULTI':[1,2],'DROP':[0]}
print('주입 준비: 유형', list(INJ_TYPES), '· σ(TRAIN 정착):', {k:round(v,2) for k,v in _SIG.items()})
```

```python
# ── P3 · 6층 평가 하네스 (계획서 §7-2 · L1~L6) ───────────────────────
from sklearn.metrics import roc_auc_score, average_precision_score
def auroc_ap(s_norm, s_anom):
    y = np.r_[np.zeros(len(s_norm)), np.ones(len(s_anom))]; s = np.r_[s_norm, s_anom]
    return float(roc_auc_score(y, s)), float(average_precision_score(y, s))
def recall_at_fpr(s_norm, s_anom, fpr=0.01):
    thr = np.quantile(s_norm, 1-fpr); return float(np.mean(np.asarray(s_anom) > thr))
def fpr_at(s_norm, thr):   return float(np.mean(np.asarray(s_norm) > thr))

class Harness:
    # score_fn: X(np.ndarray) -> wafer 이상점수(높을수록 이상). recon_fn: X -> per-feature 정규화 잔차(L1용, 선택)
    def __init__(self, name, score_fn, recon_fn=None):
        self.name, self.score_fn, self.recon_fn = name, score_fn, recon_fn
    def L1(self):  # VAL 재구성 RMSE (수렴 확인)
        if self.recon_fn is None: return None
        r = self.recon_fn(X_VAL); return float(np.sqrt((r**2).mean()))
    def L5(self, thr):  # VAL 오경보율
        return fpr_at(self.score_fn(X_VAL), thr)
    def thr_fpr1(self):  # VAL 기준 FPR1% 임계
        return float(np.quantile(self.score_fn(X_VAL), 0.99))
print('하네스 정의: Harness(L1/L2/L4/L5) · auroc_ap · recall_at_fpr — G판정 셀에서 seg2/C6_1/INJECT 로딩')
```

## P4 · 참조 베이스라인 → MLP-AE → 스코어  *(계획서 §8)*

**베이스라인 먼저**(§8-1): PCA 재구성·Mahalanobis(Ledoit-Wolf) — "AE가 이겨야 하는 선" + 하네스 디버깅을 값싼 모델로 끝냄. 이어 **MLP-AE**(§8-2) 학습, **스코어 계약**(§8-3, 단일 `ae_raw` — 불가침 12), **시드 3회 안정성**(§8-2, P5b B2 진입조건 재료).


```python
# ── P4 · 참조 베이스라인 (PCA · Mahalanobis, 계획서 §8-1) ─────────────
from sklearn.decomposition import PCA
from sklearn.covariance import LedoitWolf
pca = PCA(n_components=0.95, svd_solver='full').fit(X_TR)
def pca_score(X):                      # 채널가중 제곱 재구성 잔차 (높을수록 이상)
    Xr = pca.inverse_transform(pca.transform(X)); return (W_VEC * (X - Xr)**2).sum(1)
def pca_rmse(X):
    Xr = pca.inverse_transform(pca.transform(X)); return float(np.sqrt((W_VEC*(X-Xr)**2).mean()))
lw = LedoitWolf().fit(X_TR); _MU = X_TR.mean(0)
def maha_score(X):                     # 마할라노비스 제곱거리
    d = X - _MU; return np.einsum('ij,jk,ik->i', d, lw.precision_, d)
print(f'베이스라인: PCA {pca.n_components_}성분(95%분산) · Mahalanobis(Ledoit-Wolf)')
```

```python
# ── P4 · MLP-AE 아키텍처·학습 (계획서 §8-2) ───────────────────────────
import torch, torch.nn as nn
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
W_T = torch.tensor(W_VEC, dtype=torch.float32, device=DEVICE)
class MLPAE(nn.Module):                # D→64→32→z→32→64→D · GELU·LayerNorm·dropout
    def __init__(self, d, z=16, p=0.05):
        super().__init__()
        self.enc = nn.Sequential(nn.Linear(d,64), nn.GELU(), nn.LayerNorm(64), nn.Dropout(p),
                                 nn.Linear(64,32), nn.GELU(), nn.LayerNorm(32), nn.Linear(32,z))
        self.dec = nn.Sequential(nn.Linear(z,32), nn.GELU(), nn.LayerNorm(32),
                                 nn.Linear(32,64), nn.GELU(), nn.LayerNorm(64), nn.Linear(64,d))
    def forward(self, x): return self.dec(self.enc(x))

def train_ae(Xtr, Xval, z=16, p=0.05, lr=1e-3, denoise=0.0, max_ep=200, patience=20, seed=SEED):
    seed_all(seed)
    m = MLPAE(Xtr.shape[1], z, p).to(DEVICE)
    opt = torch.optim.AdamW(m.parameters(), lr=lr)
    sch = torch.optim.lr_scheduler.CosineAnnealingLR(opt, max_ep)
    xt = torch.tensor(Xtr, dtype=torch.float32, device=DEVICE)
    xv = torch.tensor(Xval, dtype=torch.float32, device=DEVICE)
    best, best_st, bad = 1e9, None, 0
    for ep in range(max_ep):
        m.train(); perm = torch.randperm(len(xt), device=DEVICE)
        for i in range(0, len(xt), 256):
            b = xt[perm[i:i+256]]
            inp = b + denoise*torch.randn_like(b) if denoise > 0 else b
            loss = (W_T*(m(inp)-b)**2).mean()
            opt.zero_grad(); loss.backward(); opt.step()
        sch.step(); m.eval()
        with torch.no_grad(): vl = float((W_T*(m(xv)-xv)**2).mean())
        if vl < best - 1e-6: best, best_st, bad = vl, {k:v.cpu().clone() for k,v in m.state_dict().items()}, 0
        else: bad += 1
        if bad >= patience: break
    m.load_state_dict(best_st); m.eval()
    return m, best, ep+1
print(f'MLP-AE 정의 · device={DEVICE}')
```

```python
# ── P4 · 스코어 계약 (계획서 §8-3 — 단일 ae_raw, 불가침 12) ───────────
def ae_resid(m, X):
    with torch.no_grad():
        xt = torch.tensor(X, dtype=torch.float32, device=DEVICE); return (xt - m(xt)).cpu().numpy()
def make_scorer(m):
    rv = ae_resid(m, X_VAL); mu, sd = rv.mean(0), rv.std(0) + 1e-9    # VAL 잔차분포 기준 z정규화
    def ae_raw(X, k=8):                                              # 채널가중 z제곱 top-k mean
        z = W_VEC * ((ae_resid(m, X) - mu)/sd)**2
        kk = min(k, z.shape[1]); return np.sort(z, 1)[:, -kk:].mean(1)
    def recon_rmse(X): return float(np.sqrt((W_VEC*ae_resid(m, X)**2).mean()))
    def top_channels(X, k=5):
        z = W_VEC * ((ae_resid(m, X) - mu)/sd)**2
        return [[FEAT_COLS[j] for j in row[::-1]] for row in np.argsort(z, 1)[:, -k:]]
    return {'ae_raw': ae_raw, 'rmse': recon_rmse, 'top': top_channels}
print('스코어: ae_raw(top-k z제곱) · recon_rmse(L1) · ae_top_channels(설명성)')
```

```python
# ── P4 · 채택 config + 시드 3회 재학습 (계획서 §8-2 · P5b B2 재료) ─────
CFG = dict(z=16, p=0.05, lr=1e-3, denoise=0.0)   # 기본 config (HPO는 실행 리뷰 후 §8-2, 목적함수=VAL loss만)
MODELS, SCORERS, VLS = [], [], []
for s in (42, 43, 44):
    m, vl, ne = train_ae(X_TR, X_VAL, seed=s, **CFG)
    MODELS.append(m); SCORERS.append(make_scorer(m)); VLS.append(vl)
    print(f'  seed {s}: VAL loss {vl:.5f} ({ne} epoch)')
AE = SCORERS[0]   # 대표 모델 = seed 42
print(f'시드 3회 완료 · VAL loss {np.mean(VLS):.5f}±{np.std(VLS):.5f} · config={CFG}')
```

## G1 판정 · 6층 평가  *(계획서 §7-2·§7-3 — seg2·C6_1·INJECT는 이 셀에서만 로딩)*

**불가침 2·§7-3 오염방지**: seg2·C6_1·INJECT-EVAL 피처·통계는 지금까지 **미접촉**. 이 셀에서 처음 격리 추출한다. HPO 목적함수는 VAL loss만 썼고, 이 데이터는 학습·튜닝에 닿지 않았다.
**G1 기준(§7-3)**: L2①A ≥ 0.95 · L2①B ≥ 0.90 · L2③ ≥ 0.85 · L5 ≤ 2%(P6 캘리 후 확정) · L6 PCA 대비 −0.01 이내 열세 없음 · L1 수렴. 미달 시 §6-2 피처·가중 1회전 후 재판정.


```python
# ── G1 · seg2·C6_1·INJECT-EVAL 격리 추출 + 6층 평가 ───────────────────
FEATS_S2  = extract_features(df[df['C64'].isin(SEG2)])       # 여기서 처음 로딩 (불가침 2)
FEATS_C61 = extract_features(df[df['C64'].isin(C6_1_W)])
Xs2 = transform(FEATS_S2); Xc61 = transform(FEATS_C61)
# seg2 D+0~14(전환 직후, C10 기준) vs 정착(A7 1,000장 이후)
t0 = df.loc[df['C64'].isin(SEG2), 'C10'].min()
early_wf = set(df.loc[df['C64'].isin(SEG2) & (df['C10'] <= t0 + 14*86400), 'C64'].unique())
s2_ord = df[df['C64'].isin(SEG2)].groupby('C64')['C10'].min().sort_values().index.tolist()
settle_wf = set(s2_ord[1000:])
mE = np.array([w in early_wf   for w in FEATS_S2.index]); XS2_E = Xs2[mE]
mS = np.array([w in settle_wf  for w in FEATS_S2.index]); XS2_S = Xs2[mS]

def evaluate(score_fn, recon_fn=None):
    v = score_fn(X_VAL); r = {}
    r['L2_1A'] = auroc_ap(v, score_fn(XS2_E))[0]      # VAL vs seg2 D+0~14
    r['L2_1B'] = auroc_ap(v, score_fn(XS2_S))[0]      # VAL vs seg2 정착
    r['L2_3']  = auroc_ap(v, score_fn(Xc61))[0]       # VAL vs C6_1
    for k in (0.5, 1, 2):                              # L4 저강도 주입(INJECT-EVAL seed=777)
        xa = inject_features(list(VAL_W), 'LEVEL', k, seed=777)
        r[f'L4_L{k}'] = recall_at_fpr(v, score_fn(xa))
    if recon_fn is not None: r['L1_rmse'] = recon_fn(X_VAL)
    return r

R_AE  = evaluate(AE['ae_raw'], AE['rmse'])
R_PCA = evaluate(pca_score,  pca_rmse)
R_MAHA= evaluate(maha_score)
# 시드 3회 L4 저강도 편차 (P5b B2 진입조건: >3%p)
L4S = [evaluate(sc['ae_raw'])['L4_L0.5'] for sc in SCORERS]
B2_TRIGGER = (max(L4S) - min(L4S)) > 0.03
print(f'6층 평가 완료 · 시드3회 L4@0.5 = {[round(x,3) for x in L4S]} → 편차 {max(L4S)-min(L4S):.3f} '
      f'→ B2 진입 {"발동" if B2_TRIGGER else "미발동"}')
```

```python
# ── G1 · 판정 대조표 (계획서 §7-3) ────────────────────────────────────
def _ok(v, thr, ge=True): return (v >= thr) if ge else (v <= thr)
rows = [
 ('L2①A VAL vs seg2 D+0~14', 0.95, R_AE['L2_1A'], R_PCA['L2_1A'], R_MAHA['L2_1A'], True),
 ('L2①B VAL vs seg2 정착',   0.90, R_AE['L2_1B'], R_PCA['L2_1B'], R_MAHA['L2_1B'], True),
 ('L2③ VAL vs C6_1',         0.85, R_AE['L2_3'],  R_PCA['L2_3'],  R_MAHA['L2_3'],  True),
 ('L4 LEVEL k=0.5 (sub-σ)',  None, R_AE['L4_L0.5'],R_PCA['L4_L0.5'],R_MAHA['L4_L0.5'], True),
 ('L4 LEVEL k=1',            None, R_AE['L4_L1'], R_PCA['L4_L1'], R_MAHA['L4_L1'], True),
]
tbl = []
for name, thr, ae, pca_, maha, ge in rows:
    base = max(pca_, maha)                         # L6: 베이스라인 최고 대비
    verdict = 'PASS' if (thr is None or _ok(ae, thr, ge)) else '★미달'
    l6 = 'OK' if ae >= base - 0.01 else '★열세'    # PCA/Maha 대비 −0.01 이내
    tbl.append({'지표': name, '기준': thr if thr else '(vs베이스라인)',
                'AE': round(ae,3), 'PCA': round(pca_,3), 'Maha': round(maha,3),
                'G1판정': verdict, 'L6': l6})
G1_TBL = pd.DataFrame(tbl)
L1_conv = R_AE.get('L1_rmse', np.nan)
G1_L2_PASS = all(_ok(R_AE[k], t) for k,t in [('L2_1A',0.95),('L2_1B',0.90),('L2_3',0.85)])
G1_L6_PASS = all(r['L6']=='OK' for r in tbl)
G1_PASS = bool(G1_L2_PASS and G1_L6_PASS)
show(G1_TBL)
print(f'\nL1 재구성 RMSE(VAL) = {L1_conv:.4f} (수렴 확인) · L5(오경보)는 P6 캘리 후 0.2 기준으로 확정')
print(f'G1 종합: L2 { "PASS" if G1_L2_PASS else "미달"} · L6 {"PASS" if G1_L6_PASS else "열세"} '
      f'→ **{"G1 PASS ✅ (v1 채택)" if G1_PASS else "G1 미달 → §6-2 피처·가중 1회전 후 재판정"}**')
print(f'P5b B2(앙상블) 진입조건: 시드3회 L4@0.5 편차 {max(L4S)-min(L4S):.3f} > 0.03 → {"발동" if B2_TRIGGER else "미발동"}')
# 아티팩트 저장 (채택 시 사용자가 커밋)
torch.save(MODELS[0].state_dict(), ART_DIR/'model_seed42.pt')
import pickle; pickle.dump(scaler, open(ART_DIR/'scaler.pkl','wb'))
print(f'아티팩트: {ART_DIR}/model_seed42.pt · scaler.pkl · feature_spec.json')
```
